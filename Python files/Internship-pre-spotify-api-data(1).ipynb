```python
import pandas as pd
import numpy as np
import re
```

```python
pd1 = pd.read_csv(r'spotify_millsongdata.csv')
```

```python
print(pd1[:5000]) 
```

```python
p1 = pd1['text']
#text pre processing 
#sentence split to word (tokenization)
l1 = list()
for i in p1:
    words = re.split(r'\W+', i)
    l1.append(words)
l1 = l1[:5000]
print(l1[:10])
```

```python
# only words and normalization of words to lowercase
l2 = list()
for i in l1:
    for k in i:
        if isinstance(k, str):
            al1 = 0
            for j in k:
                j = j.lower()
                if not j.isalpha():
                    al1 = 1
                    break
    if al1==0:
        l3 = list()
        for a in i:
            a = a.lower()
            l3.append(a)
        l2.append(l3)
        
print(l2[:10])
```

```python
#filtering stop words

import gensim
from gensim.parsing.preprocessing import remove_stopwords, STOPWORDS
stop_word = STOPWORDS
print(stop_word)
```

```python
l4 = list()
for i in l2:
    l5 = list()
    new_text = " ".join(i)
    new_t = remove_stopwords(new_text)
    l4.append(new_t)
print(l4[:10])
```

```python
from nltk.stem import PorterStemmer
pstemmer = PorterStemmer()
l5 = list()
for i in l4:
    words = re.split(r'\W+', i)
    l5.append(words)
```

```python
#removing smaller words 
l6 = list()
for i in l5:
    l7 = list()
    for j in i:
        if len(j)>2:
            l7.append(j)
    l6.append(l7)
```

```python
l7 = list()
for i in l6:
    l8= list()
    for j in i:
        l8.append(pstemmer.stem(j))
    l7.append(l8)
```

```python
#stemming
from nltk.stem.snowball import SnowballStemmer
snow_stem = SnowballStemmer(language='english')
```

```python
#lemmentation
from textblob import TextBlob, Word
import nltk
nltk.download('omw-1.4')
```

```python
lem = list()
for i in l7:
    l9 = list()
    for j in i:
        va = Word(j)
        val = va.lemmatize()
        l9.append(val)
    lem.append(l9)
print(lem[:10])
```

```python
sent1= list()
for i in lem:
    new_text1= " ".join(i)
    sent1.append(new_text1)
print(sent1[0])
```

```python
#feature extraction
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
lyrics_counts = vectorizer.fit_transform(sent1)
#feature_names = vectorizer.get_feature_names()
feature_names = vectorizer.vocabulary_
df_bow = pd.DataFrame(lyrics_counts.toarray(), columns = feature_names)
df_bow.head()



```

```python
from sklearn.cluster import KMeans

# Initialize the K-Means model
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit the model on the bag-of-words representation of the lyrics
kmeans.fit(df_bow)

df = pd.DataFrame()
# Assign cluster labels to each song
df['cluster'] = kmeans.labels_
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce the dimensions of the bag-of-words representation using PCA
pca = PCA(n_components=2)
lyrics_pca = pca.fit_transform(df_bow)

# Create a scatter plot of the PCA-transformed data
plt.scatter(lyrics_pca[:, 0], lyrics_pca[:, 1], c=df['cluster'], cmap='rainbow')
plt.show()
```

```python
# Combine the bag-of-words representation with the cluster labels
df_bow['cluster'] = kmeans.labels_

# Group the data by cluster
cluster_groups = df_bow.groupby('cluster')

# Find the top 20 words in each cluster
top_words = {}
for cluster, group in cluster_groups:
    top_words[cluster] = group.mean().sort_values(ascending=False).head(20).index

# Print the top words for each cluster
for cluster, words in top_words.items():
    print(f'Cluster {cluster}:\n{" ".join(words)}\n')  
```

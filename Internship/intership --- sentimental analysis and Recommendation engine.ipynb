```python
p1 = pd1['text']
#text pre processing 
#sentence split to word (tokenization)
l1 = list()
for i in p1:
    words = re.split(r'\W+', i)
    l1.append(words)
print(l1[:10])
```

```python
# only words and normalization of words to lowercase
l2 = list()
for i in l1:
    for k in i:
        if isinstance(k, str):
            al1 = 0
            for j in k:
                j = j.lower()
                if not j.isalpha():
                    al1 = 1
                    break
    if al1==0:
        l3 = list()
        for a in i:
            a = a.lower()
            l3.append(a)
        l2.append(l3)
        
print(l2[:10])
```

```python
l4 = list()
for i in l2:
    l5 = list()
    new_text = " ".join(i)
    new_t = remove_stopwords(new_text)
    l4.append(new_t)
print(l4[:10])
```

```python
from nltk.stem import PorterStemmer
pstemmer = PorterStemmer()
l5 = list()
for i in l4:
    words = re.split(r'\W+', i)
    l5.append(words)
```

```python
#removing smaller words 
l6 = list()
for i in l5:
    l7 = list()
    for j in i:
        if len(j)>2:
            l7.append(j)
    l6.append(l7)
```

```python
l7 = list()
for i in l6:
    l8= list()
    for j in i:
        l8.append(pstemmer.stem(j))
    l7.append(l8)
```

```python
#rating for each song based ortist follower number and popularity rank. 

p3 = pd.DataFrame(p2,columns=['artist','followers','popularity'])
```

```python
import math


def rating_function(followers, rank, sentiment):
    # Define a sigmoid function that maps followers to a rating between 1 and 10
    x = followers / 1000000  # Divide followers by 1 million for convenience
    rating = 1 + 9 / (1 + math.exp(-0.5*(x-1)))
    
    # Adjust the rating based on the popularity rank of the artist
    if rank == 1:
        rating *= 1.2
    elif rank <= 5:
        rating *= 1.1
    elif rank <= 10:
        rating *= 1.05
    
    # Adjust the rating based on the sentiment of the song
    if sentiment > 0.5:
        rating *= 1.2
    elif sentiment > 0.2:
        rating *= 1.1
    elif sentiment < -0.5:
        rating *= 0.8
    elif sentiment < -0.2:
        rating *= 0.9
    
    return rating
```

```python
followers =list(p3['followers'])
popularity = list(p3['popularity'])
sentiment = lis_sent
```

```python
song_rating = list()
for i in range(len(followers)):
    a = followers[i]
    b = popularity[i]
    c = sentiment[i]
    r = rating_function(a,b,c)
    song_rating.append(r)
```

```python
cosine_similarities = list(cosine_similarities4) + list(cosine_similarities5) + list(cosine_similarities6) + list(cosine_similarities7) + list(cosine_similarities8) + list(cosine_similarities9)
```

```python
p2['similarity_score'] = cosine_similarities
```

```python
def get_similar_songs(song_id, n=10):
    # Get the index of the song in the DataFrame
    idx = p2[p2['song'] == song_id].index[0]

    # Get the cosine similarities between the given song and all other songs
    sim_scores = list(enumerate(cosine_similarities[idx]))

    # Sort the songs based on the cosine similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the top N most similar songs
    sim_scores = sim_scores[1:n+1]
    song_indices = [i[0] for i in sim_scores]

    # Return the top N most similar songs
    return p2.iloc[song_indices]['song'].tolist()
```

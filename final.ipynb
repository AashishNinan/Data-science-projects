```python
import pandas as pd
import numpy as np
import math
```

```python
#updating columns of customer_details.csv
df1 = pd.read_csv('./customer_details.csv')
df1.columns = ['customer_id', 'Gender', 'age', 'driving_licence_present', 'region_code', 'previously_insured', 'vehicle_age', 'vehicle_damage'   ]
print(df1)

#updating columns of customer_policy_details.csv
df2 = pd.read_csv('customer_policy_details.csv')
df2.columns = ['customer_id', 'annual_premium', 'sales_channel_code', 'vintage', 'response']
print(df2)
```

```python
#2. Checking and Cleaning Data Quality:
#i. Null values
```

```python
#Generate a summary of count of all the null values column wise

print('This is null values in customer_details.csv')
df1_col_names = df1.columns
for i in df1_col_names:
    print(f"Null values in column {i}:  {df1[i].isnull().sum()}")

print('\n')
print('This is null values in customer_policy_details.csv')
df2_col_names = df2.columns
for i in df2_col_names:
    print(f"Null values in column {i}:  {df2[i].isnull().sum()}")
```

```python
#Drop Null values for customer_id because central tendencies for id’s is not feasible.

print('Dropping id values in column customer_id in customer_policy_details.csv')
df1 = df1.dropna(subset=['customer_id'])
#print(df1_dropna1 )
print(f"Null values in column customer_id after dropping all null value : {df1['customer_id'].isnull().sum()}")

print('\n')
print('Dropping id values in column customer_id in customer_details.csv')
df2 = df2.dropna(subset=['customer_id'])
#print(.dropna2)
print(f"Null values in column customer_id after dropping all null value : {df2['customer_id'].isnull().sum()}")
```

```python
#Replace all null values for numeric columns by mean.

print('Replacing null in numeric column with mean of customer_details.csv ')
df1_col_names = df1.columns
df1_col_types = df1.dtypes
for i in range(len(df1_col_names)):
    col_name = df1_col_names[i]
    col_type = df1_col_types[i]
    
    if col_type =='float64' or col_type =='int64':
        mean_col = round(df1[col_name].mean(),2)
        df1[col_name].fillna(value=mean_col, inplace=True)
        
for i in df1_col_names:
    print(f"Null values in column {i}:  {df1[i].isnull().sum()}")
    
print('\n')

print('Replacing null in numeric column with mean of customer_policy_details.csv ')
df2_col_names = df2.columns
df2_col_types = df2.dtypes
for i in range(len(df2_col_names)):
    col_name = df2_col_names[i]
    col_type = df2_col_types[i]
    
    if col_type =='float64' or col_type =='int64':
        mean_col = round(df2[col_name].mean(),2)
        df2[col_name].fillna(value=mean_col, inplace=True)
        
for i in df2_col_names:
    print(f"Null values in column {i}:  {df2[i].isnull().sum()}")
```

```python
#Replace all null values for Categorical value by mode.

print('Replacing null in Categorical with mode of customer_details.csv ')
for i in range(len(df1_col_names)):
    col_name = df1_col_names[i]
    col_type = df1_col_types[i]
    
    if col_type =='object':
        col_mode = df1[col_name].mode().iloc[0]
        df1[col_name].fillna(value=col_mode, inplace=True)
        
for i in df1_col_names:
    print(f"Null values in column {i}:  {df1[i].isnull().sum()}")

print('\n')

print('Replacing null in Categorical with mode of customer_policy_details.csv ')
for i in range(len(df2_col_names)):
    col_name = df2_col_names[i]
    col_type = df2_col_types[i]
    
    if col_type =='object':
        col_mode = df2[col_name].mode().iloc[0]
        df2[col_name].fillna(value=col_mode, inplace=True)
        
for i in df2_col_names:
    print(f"Null values in column {i}:  {df2[i].isnull().sum()}")
```

```python
#ii. Outliers
#Generate a summary of count of all the outliers column wise

def find_outliers_IQR(df):
    
 q1 = df.quantile(0.25)
 q3 = df.quantile(0.75)
 IQR = q3-q1
 outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
 return outliers

print('Summary of count of all the outliers column wise of customer_details.csv ')

df1_col_with_outliers = list()
for i in range(len(df1_col_names)):
    col_name = df1_col_names[i]
    col_type = df1_col_types[i]
    
    if col_type =='float64' or col_type =='int64':
        outi = find_outliers_IQR(df1[col_name])
        print(f'Number of outliers in {col_name} = {len(outi)}')
        if len(outi) :
            df1_col_with_outliers.append(col_name)
        print(f'Maximum of outliers in {col_name} = {outi.max()}')
        print(f'Minimum of outliers in {col_name} = {outi.min()}')
        print('\n')
        

print(df1_col_with_outliers)

print('Summary of count of all the outliers column wise of customer_policy_details.csv ')

print('\n')

df2_col_with_outliers = list()
for i in range(len(df2_col_names)):
    col_name = df2_col_names[i]
    col_type = df2_col_types[i]
    
    if col_type =='float64' or col_type =='int64':
        outi = find_outliers_IQR(df2[col_name])
        print(f'Number of outliers in {col_name} = {len(outi)}')
        if len(outi) :
            df2_col_with_outliers.append(col_name)
        print(f'Maximum of outliers in {col_name} = {outi.max()}')
        print(f'Minimum of outliers in {col_name} = {outi.min()}')
        print('\n')
print(df2_col_with_outliers)
```

```python
#Replace all outlier values for numeric columns by mean.

for i in range(len(df1_col_with_outliers)):
    col_name1 = df1_col_with_outliers[i]
    df = df1[col_name1]
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3-q1
    mean_col1 = round(df.mean())
    outli1 = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    lis_ot = list(outli1)
    for i in df:
        if i in lis_ot:
            df.replace(to_replace = i, value = mean_col1, inplace=True)
        
    
```

```python
#5. Is there any relation between Person Age and annual premium?
mer_age = df_merge['age']
mer_annual_premium = df_merge['annual_premium']
correlation = mer_age.corr(mer_annual_premium)

if correlation < - 0.50:
    print(f'Strong negative relationship and correlation value is {correlation}')
elif correlation > - 0.50 or correlation < 0.50:
    print(f'There is no relationship and correlation value is {correlation}')
else:
    print(f'Strong positive relationship and correlation value is {correlation}')
```
